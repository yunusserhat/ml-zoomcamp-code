```python
#used chatgpt
# Step 1: Select all the "Innjoo" laptops
innjoo_laptops = df[df['Brand'] == 'Innjoo']

# Step 2: Select only the columns RAM, Storage, Screen
X = innjoo_laptops[['RAM', 'Storage', 'Screen']].values

# Step 3: Compute matrix-matrix multiplication between the transpose of X and X
XTX = np.dot(X.T, X)

# Step 4: Compute the inverse of XTX
XTX_inv = np.linalg.inv(XTX)

# Step 5: Create an array y with values [1100, 1300, 800, 900, 1000, 1100]
y = np.array([1100, 1300, 800, 900, 1000, 1100])

# Step 6: Multiply the inverse of XTX with the transpose of X, and then multiply the result by y
w = np.dot(np.dot(XTX_inv, X.T), y)

# Step 7: Compute the sum of all elements in the result
sum_w = np.sum(w)

# Display the result
print(f"The sum of all elements in the result is: {sum_w}")
```
